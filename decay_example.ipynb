```python
from matplotlib import pyplot as plt
import numpy as np
```

This notebook will calculate the solution to the nuclear decay problem.  The change in the number of particles is proportional to the number of particles, so that

$$
\frac{dN}{dt} = -C N
$$

is the differential equation we want to solve.  We start by defining variables.

```python
steps=25
N=np.zeros(steps)
N[0]=100
C=1.3
dt=0.2
```

We start from a number of particles.  Then for each time step, we let $N = N - C N dt$.

```python
for j in range(steps-1):
    N[j+1]=N[j]-C*N[j]*dt
```

Now we can plot the results.

```python
t=np.arange(steps)*dt
plt.plot(t,N,'ro')
plt.xlabel('time')
plt.ylabel('N')
plt.show()
```

In this case, we know the analytical solution to the differential equation:

$$
N = N_0 e^{-Ct}
$$

```python
plt.plot(t,N,'ro')
plt.plot(t,N[0]*np.exp(-C*t),'k')
plt.show()
```
